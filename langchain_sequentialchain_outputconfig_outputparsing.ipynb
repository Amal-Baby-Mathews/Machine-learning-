````python
from langchain.llms import CTransformers
llm = CTransformers(model='TheBloke/Llama-2-7B-GGUF',temperature=0)

print(llm('AI is going to'))
````

````python
from langchain.prompts import PromptTemplate

from langchain.chains import LLMChain
from langchain.chains import (
    StuffDocumentsChain, ConversationalRetrievalChain
)
````

````python
article = '''Global TradeMasters, Inc. is a colossal trading organization that spans continents and industries. With a formidable presence in over 50 countries, we specialize in the seamless exchange of goods and commodities on a global scale. Our state-of-the-art trading algorithms, combined with a network of expert traders, ensure maximum profitability and minimal risk. We deal in everything from energy resources to technology components, facilitating transactions worth billions daily. Our commitment to innovation, reliability, and sustainability makes us a pioneering force in the world of international trade, fostering prosperity for nations and enterprises alike.'''
````

````python
fact_extraction_prompt = PromptTemplate(
    input_variables=["text_input"],
    template="Extract the key facts out of this text. Don't include opinions. Give each fact a number and keep them short sentences. :\n\n {text_input}"
)
````

````python
fact_extraction_chain = LLMChain(llm=llm, prompt=fact_extraction_prompt)

facts = fact_extraction_chain.run(article)

print(facts)
````

````python
triples_prompt = PromptTemplate(
    input_variables=["facts"],
    template="Take the following list of facts and turn them into triples for a knowledge graph:\n\n {facts}"
)
````

````python
triples_chain = LLMChain(llm=llm, prompt=triples_prompt)

triples = triples_chain.run(facts)

print(triples)
len(triples)
````

````python
from langchain.chains import SimpleSequentialChain, SequentialChain

full_chain = SimpleSequentialChain(chains=[fact_extraction_chain, triples_chain], verbose=True)
````

````python
response = full_chain.run(article)
````

````python
from pydantic import BaseModel, Field, validator
from typing import List
from langchain.output_parsers import PydanticOutputParser, StructuredOutputParser, ResponseSchema
````

````python
event_description="a party celebrating the independence of India"
````

````python
response_schemas = [
    ResponseSchema(name="Theme", description="Theme of the event"),
    ResponseSchema(name="score", description="Score out of ten given by you for the theme of the event. in the format: score/10")
]
output_parser = StructuredOutputParser.from_response_schemas(response_schemas)
format_instructions = output_parser.get_format_instructions()
prompt = PromptTemplate(
    template="You are a master event planner. \
You come up with Amazing event ideas according to the occasion. \ Take the event description below delimited by triple backticks and use it to create the theme for an event or party. \
event description: ```{event_description}``` then based on the description and your hot new event theme give the theme a score 1-10 for how likely it is to succeed.\n{format_instructions}"
,
    input_variables=["event_description"],
    partial_variables={"format_instructions": format_instructions}
)
_input = prompt.format_prompt(event_description="a party celebrating the independence of India")

````

````python
output = llm(_input.to_string())
````
